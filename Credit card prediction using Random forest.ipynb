# Credit card prediction using Random forest classification

## Importing the libraries

```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
```

## Importing the dataset

```python
dataset = pd.read_csv('Data/UniversalBank.csv')
```

```python
dataset.head()
```

```python
# Unwanted columns drops
dataset1 = dataset.drop(['ID','ZIP Code'],axis=1)
```

```python
#check null value columns
dataset1.isnull().sum()
```

```python
# Correlation checking
plt.figure(figsize=(15,8))
plt.title("Heatmap showing correlation between all the features", fontsize=20)
sns.heatmap(dataset1.corr(),annot = True, cmap='mako')
```

```python
sns.set(style='whitegrid')
sns.scatterplot(x="Age",
                    y="Income",
                    hue="CreditCard",
                    data=dataset1)
```

```python
sns.set(style='whitegrid')
sns.scatterplot(x="Education",
                    y="Income",
                    hue="CreditCard",
                    data=dataset1)
```

```python
sns.set(style='whitegrid')
sns.scatterplot(x="Family",
                    y="Income",
                    hue="CreditCard",
                    data=dataset1)
```

```python
sns.histplot(data=dataset1,x='Age',kde = False,hue='CreditCard')
```

## Splitting the dataset into the Training set and Test set

```python
X = dataset1.iloc[:, :-1].values
y = dataset1.iloc[:, -1].values
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.25, random_state = 0)
```

```python
print(X_train)
```

## Feature Scaling

```python
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
X_train = sc.fit_transform(X_train)
X_test = sc.transform(X_test)
```

```python
print(X_train[:1])
```

```python
print(X_train)
```

```python
print(X_test)
```

## Training the random forest model on the Training set

```python
from sklearn.ensemble import RandomForestClassifier
classifier = RandomForestClassifier(n_estimators = 10, criterion = 'entropy', random_state = 0)
classifier.fit(X_train, y_train)
```

## Predicting the Test set results

```python
y_pred = classifier.predict(X_test)
print(np.concatenate((y_pred.reshape(len(y_pred),1), y_test.reshape(len(y_test),1)),1))
```

## Making the Confusion Matrix

```python
from sklearn.metrics import confusion_matrix, accuracy_score
cm = confusion_matrix(y_test, y_pred)
print(cm)
accuracy_score(y_test, y_pred)
```
